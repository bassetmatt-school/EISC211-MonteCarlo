```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

modif = {"font.family" : "serif", "font.serif" : "CMU Serif"}

matplotlib.rcParams["font.family"] = "serif"
matplotlib.rcParams["font.serif" ] = "CMU Serif"

matplotlib.rcParams["text.usetex"] = "true"

###
# Actual code

fig = plt.figure()

fig.suptitle("Title")
plt.xlabel(r"Label $x_i$")
plt.ylabel(r"Label $y_i$")

X = np.linspace(0,10,1000)
Y1 = np.power(X,2)
Y2 = 80 - 5*X

plt.plot(X,Y1)
plt.plot(X,Y2)

plt.show()
```
